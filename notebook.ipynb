```python
import numpy as np
import matplotlib.pyplot as plt

# Simple test computation
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave Demo for Binder")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()

print("âœ… Notebook is running fine on Binder!")
```
